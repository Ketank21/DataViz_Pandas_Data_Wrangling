## 4) Melting dataframes into long formats

```python
df_fert =df_fert.melt(id_vars=['country'],
                     var_name='year',
                     value_name ='fertility_rate')
df_fert.head()
```

```python
df_LE =df_LE.melt(id_vars=['country'],
                     var_name='year',
                     value_name ='life_expectancy')
df_LE.head()
```

```python
df_pop =df_pop.melt(id_vars=['country'],
                     var_name='year',
                     value_name ='total_population')
df_pop.head()
```

## 5) Merge Dataframes

```python
df_gapminder_merged =pd.merge(df_fert,df_LE,on =['country','year'],how ='inner')
df_gapminder_merged.head()
```

```python
df_gapminder_merged =pd.merge(df_gapminder_merged,df_pop,on =['country','year'],how = 'left')
df_gapminder_merged.head()
```

```python
df_gapminder_merged = pd.merge(df_gapminder_merged,df_cont,on ='country',how='left')
df_gapminder_merged.head()
```

## Plotting Graphs from the data frame

```python
#https://www.python-graph-gallery.com/bubble-plot/       #used this a lot!!!
import seaborn as sns
import matplotlib.pyplot as plt
for year in range(1960,2015):

    plt.figure(figsize=(10,6))
    df_subset=df_gapminder_merged.loc[df_gapminder_merged['year']==year]
    sns.scatterplot(x='life_expectancy',y='fertility_rate',
                    data=df_subset,size="total_population",
                    alpha=0.6,legend="brief",sizes=(20,2000),
                    hue='continent')
    s = str(year)
    plt.title('Fertility rate vs Life Expectency from 1960 to 2015',fontsize=14)
    plt.xlabel("Life expectancy / years",loc= "right",fontsize= 12)
    plt.ylabel("Fertility rate / %", loc="top",fontsize=12)
    plt.text(30,9,s,fontsize = 16)#https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.text.html
    plt.axis(xmin=20,xmax =100,ymin=0,ymax=10)
    plt.legend(loc=(0.85,0.5))
    sns.despine()
    plt.savefig("Graphs/lifeexp_"+ str(year) +".png")
    year =year + 1
    plt.close()
    
```

```python
images=[]

for i in range(1960,2015):
    filename ='Graphs/lifeexp_{}.png'.format(i)
    images.append(imageio.imread(filename))
imageio.mimsave('output.gif',images,fps=5)
```
